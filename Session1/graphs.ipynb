```python
import folium
import pandas as pd

tourist_attractions = pd.read_csv("tourist_attractions.csv")
```

```python
tourist_attractions
```

```python
center = [21.3279755, -157.9395024]
map_hawaii = folium.Map(location=center, zoom_start=8)
#display map
map_hawaii
```

```python
for index, tourist_attraction in tourist_attractions.iterrows():
    location = [float(tourist_attraction['lat']), float(tourist_attraction['long'])]
    folium.Marker(location, popup = f'Name:{tourist_attraction["title"]}').add_to(map_hawaii)
    
map_hawaii
```

```python
hotels = pd.read_csv("modified_dataset_expedia.csv")
```

```python
center = [21.3279755, -157.9395024]
map_hotels = folium.Map(location=center, zoom_start=8)

for index, hotel in hotels.iterrows():
    location = [float(hotel['latitude']), float(hotel['longitude'])]
    folium.Marker(location, popup = f'Name:{hotels["name"]}').add_to(map_hotels)
    
map_hotels
```

```python
hw_properties = pd.read_csv("HawaiiProperties.csv")
```

```python
center = [21.3279755, -157.9395024]
map_properties = folium.Map(location=center, zoom_start=8)

for index, properties in hw_properties.iterrows():
    location = [float(properties['latitude']), float(properties['longitude'])]
    folium.Marker(location).add_to(map_properties)
    
map_properties
```

```python
df_airbnb = pd.read_csv("Airbnb_ratio.csv")[["latitude", "longitude"]]
lst_airbnb = list(list(tup[1:]) for tup in df_airbnb.itertuples())
lst_airbnb[:5]
```

```python
df_airbnb = pd.read_csv("Airbnb_ratio.csv")[["latitude", "longitude", "price"]]
df_airbnb["price"] = (df_airbnb["price"]-1)
lst_airbnb_price = list(list(tup[1:]) for tup in df_airbnb.itertuples())
lst_airbnb_price[:5]
```

```python
from folium.plugins import HeatMap

center = [21.4079755, -157.9395024]
heatmap = folium.Map(location=center, tiles='cartodbpositron', zoom_start=10)
HeatMap(lst_airbnb, "airbnb", radius=15, blur=10, gradient={.7:'blue',.95:'lime',.98:'red'}).add_to(heatmap)
heatmap
```

```python
heatmap_2 = folium.Map(location=center, tiles='cartodbpositron', zoom_start=10)
HeatMap(lst_airbnb_price, "airbnb", radius=15, blur=10, gradient={.7:'blue',.95:'lime',.98:'red'}).add_to(heatmap_2)
heatmap_2
```
